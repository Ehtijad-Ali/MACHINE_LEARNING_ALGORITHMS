```python
inputs= df.drop('salary', axis='columns')
target=df['salary']
target
```

```python
print(df.columns)
```

```python
from sklearn.preprocessing import LabelEncoder

le_work_year=LabelEncoder()
le_salary_in_usd=LabelEncoder()
le_remote_ratio=LabelEncoder()


```

```python
inputs["work_year_n"]=le_work_year.fit_transform(inputs['work_year'])
inputs["salary_in_usd_n"] = le_work_year.fit_transform(df["salary_in_usd"])
inputs["remote_ratio_n"]=le_work_year.fit_transform(inputs["remote_ratio"])
inputs.head(3)
```

```python
inputs_n=inputs.drop(['job_title','employment_type','experience_level','salary_currency','employee_residence','company_location','company_size'],axis='columns')
inputs_n
```

```python
from sklearn import tree
model=tree.DecisionTreeClassifier()
model.fit(inputs_n, target)
```

```python
model.score(inputs_n, target)
```
