```python
df['Day'] = df['Date_of_Journey'].str.split('/').str[0]
df['Month'] = df['Date_of_Journey'].str.split('/').str[1]
df['Year'] = df['Date_of_Journey'].str.split('/').str[2]
```

```python
df.head()
```

```python
df['Day'] = df['Day'].astype(int)
df['Month'] = df['Month'].astype(int)
df['Year'] = df['Year'].astype(int)
```

```python
df.head(2)
```

```python
df.drop('Date_of_Journey', axis = 1, inplace = True)
```

```python
df.info()
```

```python
df.head(10)
```

```python
df['Arrival_Time'].str.split(' ').str[0]
```

```python
df['Arrival_Hour'] = df['Arrival_Time'].str.split(' ').str[0].str.split(':').str[0]
df['Arrival_Minute'] = df['Arrival_Time'].str.split(' ').str[0].str.split(':').str[1]
df.drop('Arrival_Time', axis = 1, inplace = True)
```

```python
df.info()
```

```python
df['Arrival_Hour'] = df['Arrival_Hour'].astype(int)
df['Arrival_Minute'] = df['Arrival_Minute'].astype(int)
```

```python
df.info()
```

```python
df.head()
```

```python
df['Dep_Hour'] = df['Dep_Time'].str.split(':').str[0].astype(int)
df['Dep_Minute'] = df['Dep_Time'].str.split(':').str[1].astype(int)
df.drop('Dep_Time', axis = 1, inplace = True)
```

```python
df.info()
```

```python
df.head()
```

```python
df.drop('Route', axis = 1, inplace = True)
```

```python
df.head(2)
```

```python
df['Duration_Hour'] = df['Duration'].str.split(' ').str[0].str.split('h').str[0]
df['Duration_Minute'] = df['Duration'].str.split(' ').str[1].str.split('m').str[0]
df.drop('Duration', axis = 1, inplace = True)
```
